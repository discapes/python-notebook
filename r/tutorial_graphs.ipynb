## R Graphics

```r
x <- c(1, 2, 3, 4, 5)
y <- c(3, 7, 8, 9, 12)

plot(5:10, main="My Graph", xlab="The x-axis", ylab="The y axis", col="red")
plot(x, y, cex=4, pch=5) # cex = size, pch = shape
plot(x, y, type="l", lwd=3, lty=5) # line width, line style
```

pch:

![shapes](https://www.w3schools.com/r/r_plot_pch2.png)

lty:

        0 removes the line
        1 displays a solid line
        2 displays a dashed line
        3 displays a dotted line
        4 displays a "dot dashed" line
        5 displays a "long dashed" line
        6 displays a "two dashed" line

```r
# To display more than one line in a graph, use the plot() function together with the lines() function:
line1 <- c(1,2,3,4,5,10)
line2 <- c(2,5,7,8,9,10)

plot(line1, type = "l", col = "blue")
lines(line2, type="l", col = "red")
points(line2, type="l", col = "red")
# apparently lines() and points() do the same thing, but points() is meant for scatter plots?
```

## Pie 

```r
x <- c(10,20,30,40)
pie(x)
# Display the pie chart and start the first pie at 90 degrees
pie(x, init.angle = 90)
```

```r
labels <- c("Apples", "Bananas", "Cherries", "Dates")
colors <- c("blue", "yellow", "green", "black")

# Display the pie chart with labels
pie(x, label = labels, main = "Fruits", col = colors)
# Display the explanation box
legend("bottomright", labels, fill = colors)
```
